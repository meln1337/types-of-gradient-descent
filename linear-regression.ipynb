```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def line(k, x, b):
    return k * x + b
```

```python
class LinearRegression:
    def __init__(self):
        self.intercept = 0
        self.slope = 1
        self.learning_rate = 0.000001
    def fit(self, training_inputs, training_outputs, epochs=1000, debug=True):
        self.training_inputs = training_inputs
        self.training_outputs = training_outputs
        self.epochs = epochs
        self.debug = debug
    def gradient_intercept(self):
        return np.sum(-2 * (self.training_outputs - line(self.slope, self.training_inputs, self.intercept)))
    def gradient_slope(self):
        return np.sum(2 * -self.training_inputs * (self.training_outputs - line(self.slope, self.training_inputs, self.intercept)))
    def loss(self):
        pred = self.training_inputs * self.slope + self.intercept
        return np.sum((self.training_outputs - pred) ** 2)
    def train(self):
        self.learning_rate_intercept = 10 ** -(len(str(self.gradient_intercept())))
        self.learning_rate_slope = 10 ** -(len(str(self.gradient_slope())))
        if self.debug == True:
            print("Loss before training:", self.loss())
            print("Intercept before training:", self.intercept)
            print("Slope before training:", self.slope)
            for i in range(self.epochs):
                self.d_intercept = self.gradient_intercept() * 0.01
                self.d_slope = self.gradient_slope() * 0.00000000001
                self.intercept -= self.d_intercept
                self.slope -= self.d_slope
                print("Iteration:", i)
                print("Intercept:", self.intercept)
                print("Slope:", self.slope)
                print("Loss:", self.loss())
                print("")
        else:
            for i in range(self.epochs):
                self.d_intercept = self.gradient_intercept() * 0.1
                self.d_slope = self.gradient_slope() * 0.00000000001
                self.intercept -= self.d_intercept
                self.slope -= self.d_slope
        print("Intercept after training")
        print(self.intercept)
        print("Slope after training")
        print(self.slope)
        print("Loss after training")
        print(self.loss())
        return self.intercept, self.slope
```

```python
x = np.arange(150,210)
training_x = np.array([175,180,150,190,200,178,170,160])
training_y = np.array([60,75,40,85,90,73,58,46])

model = LinearRegression()
model.fit(training_x, training_y, debug=True)
intercept, slope = model.train()
plt.plot(training_x, training_y, '.', x, line(slope, x, intercept))
```
